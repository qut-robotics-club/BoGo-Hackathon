```python
import requests
import shutil

# Set up the endpoint, headers, and data
url = 'https://api.topmediai.com/v1/text2speech'
headers = {
    'accept': 'application/json',
    'x-api-key': '51867a2bc5d640378761efc117186f78',  # Replace with your own API key
    'Content-Type': 'application/json'
}
data = {
    "text": "My name is Biden, your really fluffy",
    "speaker": "00151130-3826-11ee-a861-00163e2ac61b",
    "emotion": "Neutral"
}

# Make the POST request
response = requests.post(url, headers=headers, json=data)

# Check for a valid response
if response.status_code == 200:
    # Parse the JSON response
    response_data = response.json()
    
    # Get the URL of the audio file
    audio_url = response_data['data']['oss_url']
    
    # Download the audio file
    audio_response = requests.get(audio_url, stream=True)
    
    # Check for a valid response for audio file
    if audio_response.status_code == 200:
        # Save the audio file to disk
        with open('output.wav', 'wb') as audio_file:
            audio_response.raw.decode_content = True
            shutil.copyfileobj(audio_response.raw, audio_file)
        print('Audio file saved as output.wav')
    else:
        print(f'Failed to retrieve audio file: {audio_response.status_code}')
else:
    print(f'Failed to convert text to speech: {response.status_code}')
```
